```python
import pandas as pd
```

```python
file_name = 'student-mat.csv'
data_frame = pd.read_csv(file_name, delimiter=';')
```

```python
data_frame.head(10)
```

```python
#Modify the DataFrame to contain only these columns: school , sex , age ,
#address , health , absences , G1 , G2 , and G3 :
data_frame = data_frame[['school', 'sex', 'age', 'address', 'health', 'absences', 'G1', 'G2', 'G3']]
```

```python
data_frame.head(10)
```

```python
#Build a pivot table indexed on 'school'
pd.pivot_table(data_frame,index=['school'])
```

```python
#pivot table indexed on 'school' and 'age' :
pd.pivot_table(data_frame, index=['school', 'age'])
```

```python
pd.pivot_table(data_frame, index=['school', 'sex', 'age'], values = 'absences', aggfunc={'mean', 'sum'})
```
