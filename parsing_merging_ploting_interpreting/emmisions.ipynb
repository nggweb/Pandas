```python
import pandas as pd
from dateutil.parser import parse
```

```python
filePath = 'AirQualityUCI.csv'

data= pd.read_csv(filePath,delimiter=';')
data.head()
```

```python
data = data.drop(['Unnamed: 15', 'Unnamed: 16'],axis=1)
data.head()
```

```python
data.shape
```

```python
data = data.dropna()
data.shape
```

```python
data.info()
```

```python
# parsing the date
data['Parse_date'] = data['Date'].apply(lambda x:parse(x))

# parsing the weekday
data['Weekday'] = data['Parse_date'].apply(lambda x:x.weekday())

#Parsing the DAY
data['Day'] = data['Parse_date'].apply(lambda x: x.strftime("%A"))

#Parsing the Month

data['Month'] = data['Parse_date'].apply(lambda x: x.strftime("%B"))

data.head()
```

```python
## Which day of the week has the highest NO2(GT) emissions?
```

```python
#Aggregate NO2(GT) based on Day and find the mean emissions for the days

Q1_1 = pd.DataFrame(data.groupby(['Day'])['NO2(GT)'].agg('mean'))
Q1_1
```

```python
# Make a bar plot with the data

DayPlot = Q1_1.plot.bar(y='NO2(GT)',rot=90,title='Emissions on week days')
```

```python
#what time of the day are NMHC(GT) emissions highest?
```

```python
## Aggregate the data based on the time column and find the mean of 'S

Q2_1 = pd.DataFrame(data.groupby(['Time'])['NMHC(GT)'].agg('mean'))
Q2_1
```

```python
# Horizontal bar plot

timePlot = Q2_1.plot.barh(y= 'NMHC(GT)',title = 'Emissions at hours of day')

timePlot.set_xlabel("Hours")
timePlot.set_ylabel("Average Emissions")
```

```python
# Which month has the highest CO(GT) emissions?
```

```python
# Creating the function to clean
def cleanFeat(x):
    return pd.to_numeric(".".join(x.split(',')))
```

```python
# Cleaning up the CO format

data['CO(GT)'] = data['CO(GT)'].apply(lambda x:cleanFeat(x))
data.head()
```

```python
## Aggregate the data based on the Month and find the mean\
#of 'CO(GT)' column

Q3_1 = pd.DataFrame(data.groupby(['Month'])['CO(GT)'].agg('mean'))
Q3_1

```

```python
# Plot the data
monthPlot = Q3_1.plot.barh(y= 'CO(GT)',title='Monthly average emissions')

monthPlot.set_xlabel("Months")
monthPlot.set_ylabel("Average Emissions")
```
